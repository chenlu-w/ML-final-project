```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Sleep_Efficiency.csv')
data = data.dropna()
```

```python
data.describe()
```

### Scatter Plot

```python
sns.scatterplot(data=data, x="Age", y="Sleep efficiency", hue="Age")
plt.show()
```

### Correlation heapmap

```python
# Select only numerical columns
numerical_data = data.select_dtypes(include=[np.number]).drop('ID', axis=1)

# Calculate the correlation matrix with only the numerical data
correlation_matrix = numerical_data.corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))  # Adjust the size to your preference
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.show()
```
